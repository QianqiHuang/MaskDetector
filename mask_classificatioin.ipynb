```
import torch
from PIL import Image
from torchvision import transforms as trn
from torch.autograd import Variable as V
import pandas as pd
device = torch.device('cuda')


def preprocess(image_fname):
    center_crop = trn.Compose([
        trn.Resize((256, 256)),
        trn.CenterCrop(224),
        trn.ToTensor(),
        trn.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])
    img_input = Image.open(image_fname)
    return V(center_crop(img_input).unsqueeze(0))

    
def prediect(img_path):
    
    net=torch.load('gdrive/My Drive/mask_detection/googlenet_masknet.pth')
    #net.cuda()
    #net.load_state_dict(torch.load('gdrive/My Drive/mask_detection/mobilenet_v2_masknet.pth'))
    # net=net.to(device)
    # torch.no_grad()
    # img=Image.open(img_path)
    # img=transform(img).unsqueeze(0)
    img=preprocess(img_path)
    # img_ = img.to(device)
    # outputs = net(img_)
    img=img.type(torch.cuda.FloatTensor)
    outputs = net(img)
    _, predicted = torch.max(outputs, 1)
    # print(predicted)
    print('this picture maybe :',predicted[0])
    return predicted[0]

def test():
  data_path = 'gdrive/My Drive/mask_detection/dataset/'
  test = pd.read_csv((data_path + 'test/test.csv'), index_col=0)
  cor = 0
  err = 0
  for i in range(test.shape[0]):
    filename = test.iloc[i, 0]
    label = test.iloc[i, 1]
    rlt = prediect('gdrive/My Drive/mask_detection/dataset/test/' + filename)
    if(rlt == label):
      cor += 1.0
    else:
      err += 1.0
    AP = cor / (cor + err)
    print(AP)



if __name__ == '__main__':
    #prediect('gdrive/My Drive/mask_detection/dataset/test/01061.png')
    test()
```
